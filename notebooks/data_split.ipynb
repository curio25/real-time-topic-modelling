```python
import pandas as pd
```

```python
file_path = '../data/netflix_reviews.csv'
df = pd.read_csv(file_path)
df.info()
```

```python
# Convert the 'at' column to datetime format
df['at'] = pd.to_datetime(df['at'])

# Split the data into historical (2018-2023) and live (2024)
historical_data = df[df['at'].dt.year < 2024]
live_data = df[df['at'].dt.year == 2024]
```
